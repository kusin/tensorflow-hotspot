### Visualisasi Data Karhutla 2014-2019

```python
data ={
    "tahun":["2014","2015","2016","2017","2018","2019"],
    "luas":[44411, 2611411, 438363, 165484, 529267, 1649258]
}

df = pd.DataFrame(data)
df
```

```python
## membuat kanvas
fig, ax = plt.subplots(num=None, figsize=(7, 4))

labels = df["tahun"]
x = np.arange(len(labels))

## membuat bar plot
bar = ax.bar(df["tahun"], df["luas"], color = "#FFC222", alpha=1, label="Luas Area (Ha)")

# membuat label-label
ax.set_xlabel('Tahun Kejadian', fontsize=12)
ax.set_ylabel('Luas Area', fontsize=12)
ax.set_ylabel('Luas Area', fontsize=12)

# limit nilai y
ax.set_ylim([0, 3_000_000])
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
#plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}')) # No decimal places

# value bar plot
ax.set_xticks(x)
ax.set_xticklabels(labels)

# membuat label xtick
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate("{:,}".format(int(height)), xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 7), # 3 points vertical offset
                    textcoords="offset points", ha='center', va='bottom')

# memanggil label
autolabel(bar)

# Menampilkan hasil xtick
fig.tight_layout()

ax.legend(loc='upper right')
ax.grid(True)

# Menampilkan barplot
plt.show()
```

```python
# set width of bar
barWidth = 0.15

# inputan dataset
t_2017 = [40, 65, 39, 50, 99, 91, 350, 539, 855, 507, 106, 72]
t_2018 = [74, 219, 75, 86, 67, 133, 613, 3090, 2353, 1411, 255, 83]
t_2019 = [49, 275, 396, 155, 172, 133, 592, 3428, 16178, 4304, 1874, 201]
t_2020 = [90, 156, 252, 155, 66, 59, 114, 945, 445, 420, 164, 0]


# Set position of bar on X axis
r1 = np.arange(len(t_2017))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

# membuat bingkai
fig, ax = plt.subplots(num=None, figsize=(10, 5))

# membuat barplot
ax.bar(r1, t_2017, width=barWidth, label='Tahun 2017', edgecolor='white', color='red')
ax.bar(r2, t_2018, width=barWidth, label='Tahun 2018', edgecolor='white', color='orange')
ax.bar(r3, t_2019, width=barWidth, label='Tahun 2019', edgecolor='white', color='green')
ax.bar(r4, t_2020, width=barWidth, label='Tahun 2020', color='blue',  edgecolor='white')

# membuat label x dan y
plt.xlabel("Bulan", fontsize=12)
plt.ylabel("Jumlah Titik Panas", fontsize=12)

# membuat nilai x
plt.xticks([r + barWidth for r in range(len(t_2020))], ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Ags', 'Sep', 'Okt', 'Nov', 'Des'])

# mengatur format nilai y
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

# membuat grid
plt.grid(True)

# Create legend & Show graphic
plt.legend()
```
