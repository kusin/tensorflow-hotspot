### Normalisasi data

```python
# membaca dataset
df_norm = read_excel("dataset/dataset.xlsx")
```

```python
# set index tanggal
df_norm = dataset.set_index("tanggal")
```

```python
print(df_norm)
```

```python
# memanggil fungsi max min scaler
scaler = preprocessing.MinMaxScaler(feature_range=(-1, 1))
```

```python
# proses max min scaler
df_norm = scaler.fit_transform(df_norm)
```

```python
df_norm = pd.DataFrame(df_norm, columns=['sst', 'soi', 'oni', 'curah_hujan', 'hotspot_sumsel'])
```

```python
print(df_norm)
```
